### 이미지로 만들기

```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

k = 500
z = [i for i in range(k)]
x = [math.cos(i/10)*(k-i) for i in range(k)]
y = [math.sin(i/10)*(k-i) for i in range(k)]
ax.scatter(x, y, z, c = 'green', marker = '*')

k = 500
z = [i for i in range(1, k, 4)]
x = [math.cos(i/10)*(k-i+10) for i in range(1, k, 4)]
y = [math.sin(i/10)*(k-i+10) for i in range(1, k, 4)]
ax.scatter(x, y, z, c = 'red', s = 10, marker = '*')
ax.scatter(x[-1], y[-1], z[-1], c = 'yellow', s = 600, marker = '*')
ax.scatter(x[-1], y[-1], z[-1], c = 'red', s = 300, marker = '*')



z = np.random.randint(0, 400, 100)
x = np.random.randint(-1000, 1000, 100)
y = np.random.randint(-1000, 1000, 100)
ax.scatter(x, y, z, c= 'gray')

z = np.random.randint(400, 500, 10)
x = np.random.randint(-1000, 1000, 10)
y = np.random.randint(-1000, 1000, 10)
ax.scatter(x, y, z, c= 'yellow')

plt.xlim(-1000, 1000)
plt.ylim(-1000, 1000)
plt.title('Merry Christmas ☆', color = 'r', fontsize = 20, position=(0.52, -100))
plt.axis('off')
plt.savefig('Merry Christmas.png', facecolor = 'black')
plt.show()
```

### gjf로 만들기

```python
import matplotlib.pyplot as plt
from matplotlib import gridspec

import seaborn as sns
import celluloid
import numpy as np
import scipy.stats as stats
import math
from mpl_toolkits.mplot3d import axes3d

from moviepy.editor import *


fig, axes = plt.subplots()
fig.patch.set_facecolor('black')
camera = celluloid.Camera(fig)    
a = np.random.randint(500, 1000, 200)
b = np.random.randint(-1000, 1000, 200)
c = np.random.randint(-1000, 1000, 200)
A = np.random.randint(500, 1000, 50)
B = np.random.randint(-1000, 1000, 50)
C = np.random.randint(-1000, 1000, 50)
for i in range(100):

    ax = plt.subplot(projection = '3d')
    ax.set_facecolor('black')
    k = 500
    z = [i for i in range(0, k*2, 2)]
    x = [math.cos(i/10)*(k-i) for i in range(k)]
    y = [math.sin(i/10)*(k-i) for i in range(k)]
    ax.scatter(x, y, z, c = 'green', marker = '*')

    k = 500
    z = [i for i in range(1, k*2, 8)]
    x = [math.cos(i/10)*(k-i+10) for i in range(1, k, 4)]
    y = [math.sin(i/10)*(k-i+10) for i in range(1, k, 4)]
    ax.scatter(x, y, z, c = 'red', s = 10, marker = '*')
    
    if i % 2 == 0:
        ax.scatter(x[-1], y[-1], z[-1], c = 'yellow', s = 600, marker = '*')
    else:
        ax.scatter(x[-1], y[-1], z[-1], c = 'white', s = 600, marker = '*')
    ax.scatter(x[-1], y[-1], z[-1], c = 'red', s = 300, marker = '*')

    z = a
    x = b
    y = c
    ax.scatter(x, y, z, c= 'white')
    if i % 2 == 0:
        z = A
        x = B
        y = C
        ax.scatter(x, y, z, c= 'yellow', s = .7, marker = '*')
    else:
        pass

    plt.xlim(-1000, 1000)
    plt.ylim(-1000, 1000)
    plt.title('Merry Christmas ☆', color = 'r', fontsize = 20, position=(0.52, -100))
    plt.axis('off')
    plt.savefig('Merry Christmas.png', facecolor = 'black')
    # plt.show()
    
    camera.snap()
    a = a*0.98

animation = camera.animate()
animation.save("gamma_alpha.mp4")

clip = (VideoFileClip("gamma_alpha.mp4").speedx(5))
clip.write_gif("gamma_alpha.gif")
```
